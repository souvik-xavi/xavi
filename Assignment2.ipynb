```
import numpy as np
import matplotlib.pyplot as plt

ac=float(input("Enter the Amplitude for Carrier"))
fc=float(input("Enter the Frequency for Carrier"))
am=float(input("Enter the Amplitude for Signal"))
fm=float(input("Enter the frequency for Signal"))

t=np.linspace(0,1,1000)
m=am*np.cos(2*np.pi*fm*t)
c=ac*np.cos(2*np.pi*fc*t)
AM=(ac+m)*np.cos(2*np.pi*fc*t)

%matplotlib inline
plt.subplot(3,1,1)
plt.title("Amplitude Modulation")
plt.plot(m,'g')
plt.ylabel("Amplitude")
plt.xlabel("Signal")

plt.subplot(3,1,2)
plt.plot(c,'r')
plt.ylabel("Amplitude")
plt.xlabel("Carrier")

plt.subplot(3,1,3)
plt.plot(AM,'b')
plt.ylabel("Amplitude")
plt.xlabel("Amplitude Modulated Signal")



plt.subplots_adjust(hspace=1)
plt.rc('font',size=15)
fig=plt.gcf()
fig.set_size_inches(16,9)
fig.savefig("Amplitude Modulation.png",dpi=100)
```
