```python
import numpy as np
import matplotlib.pyplot as plt
ac=float(input("Enter the amplitude of carrier:"))
fc=float(input("Enter the frequency of carrier:"))
am=float(input("Enter the amplitude of msg:"))
fm=float(input("Enter the frequency of msg:"))
t=np.linspace(0,1,1000)
m=am*np.cos(2*np.pi*fm*t)
c=ac*np.cos(2*np.pi*fc*t)
FM=ac*np.cos(2*np.pi*fc*t + m)
```

```python
plt.subplot(3,1,1)
plt.title('Frequency Modulation')
plt.plot(m,'g')
plt.ylabel('Message')
plt.xlabel('Frequency')
```

```python
plt.subplot(3,1,2)
plt.plot(c,'r')
plt.ylabel('Carrier')
plt.xlabel('Frequency')
```

```python
plt.subplot(3,1,3)
plt.plot(FM,'g')
plt.ylabel('Modulated Signal')
plt.xlabel('Frequency')
```

```python
plt.subplots_adjust(hspace=1)
plt.rc('font',size=15)
fig=plt.gcf()
fig.set_size_inches(16,9)
fig.savefig('Frequency Modulation.png',dpi=100)
```
